```python
import pandas as pd
import matplotlib.pyplot as plt

# Create Retail Sales Dataset
data = {
    "Product": ["Laptop","Mobile","Shoes","Shirt","Rice","Oil","TV","Watch","Bag","Book"],
    "Category": ["Electronics","Electronics","Fashion","Fashion","Grocery","Grocery","Electronics","Accessories","Accessories","Stationery"],
    "Quantity": [5,10,20,15,30,25,3,12,18,40],
    "Price": [50000,20000,2500,1200,60,180,45000,3500,1500,300]
}

df = pd.DataFrame(data)

# Create Total Sales Column
df["TotalSales"] = df["Quantity"] * df["Price"]

print("First Five Rows")
print(df.head())

print("\nStatistical Summary")
print(df.describe())

print("\nCategory Wise Sales")
category_sales = df.groupby("Category")["TotalSales"].sum()
print(category_sales)

# -------------------------
# Bar Chart
# -------------------------
plt.figure(figsize=(8,5))
category_sales.plot(kind="bar")
plt.title("Category Wise Sales")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.grid(axis='y')
plt.show()

# -------------------------
# Pie Chart
# -------------------------
plt.figure(figsize=(6,6))
category_sales.plot(kind="pie", autopct="%1.1f%%")
plt.title("Sales Distribution")
plt.ylabel("")
plt.show()

# -------------------------
# Product Sales
# -------------------------
plt.figure(figsize=(8,5))
plt.bar(df["Product"], df["TotalSales"])
plt.xticks(rotation=45)
plt.title("Product Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()

# -------------------------
# Top Selling Product
# -------------------------
top_product = df.loc[df["TotalSales"].idxmax()]

print("\nTop Selling Product")
print(top_product)

print("\nConclusion")
print("1. Electronics generated the highest revenue.")
print("2. Grocery products sold the highest quantity.")
print("3. Laptops contributed the highest sales value.")
print("4. Retail businesses should focus on Electronics for maximum profit.")
```
